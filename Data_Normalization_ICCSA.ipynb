```python
def df_to_row(df):
    df_new = pd.DataFrame([{
        'Date': pd.to_datetime(df['time']).dt.strftime('%Y-%m-%d').iloc[0],
        'Weight Mean': df['weight'].mean(),
        'Difficulty Mean': df['difficulty'].mean(),
        'Reward Mean': df['reward_usd'].mean(),
        'Transaction Sum': df['transaction_count'].sum(),
        'Witness Sum': df['witness_count'].sum(),
        'Input Sum': df['input_count'].sum(),
        'Output Sum': df['output_count'].sum(),
        'Fee Total Sum': df['fee_total_usd'].sum(),
        'Total Blocks': len(df)
    }])

    return df_new
```

```python
def download_file(url, local_filename):
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(local_filename, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                f.write(chunk)
    return local_filename
```

```python
import glob
import os

def listar_arquivos(caminho_da_pasta):
    arquivos = glob.glob(os.path.join(caminho_da_pasta, '**'), recursive=True)
    return [arquivo for arquivo in arquivos if os.path.isfile(arquivo)]

# Usage example:
caminho = '/content/downloads/'
arquivos = listar_arquivos(caminho)

print(arquivos)
```

```python
import os
import pandas as pd

# Folder where the CSV files are
caminho_pasta = "/content/bitcoin_coinmarketcap"

# List to store DataFrames
dfs = []

# Browse all files in the folder
for arquivo in os.listdir(caminho_pasta):
    if arquivo.endswith(".csv"):  # Make sure they are CSV files
        caminho_arquivo = os.path.join(caminho_pasta, arquivo)

        # Try different delimiters (if it's ";" instead of ","")
        try:
            df = pd.read_csv(caminho_arquivo, sep=";", encoding="utf-8")
        except:
            df = pd.read_csv(caminho_arquivo, sep=",", encoding="utf-8")

        df = df.drop(['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'name'],axis=1)
        # Normalize the date column
        # Convert timestamp column to datetime
        df['timestamp'] = pd.to_datetime(df['timestamp'])

        # Round to nearest day
        df['timestamp'] = df['timestamp'].dt.round('D').dt.date  # Retorna um objeto datetime.date

        # Add to list
        dfs.append(df)

# Concatenate all DataFrames ordering by date
df_final = pd.concat(dfs).sort_values(by="timestamp").reset_index(drop=True)

# Display first lines
df_final.head()
```

```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit

# Logarithmic function used in adjustment
def log_func(x, a, b, c):
    return a * np.log(b + x) + c

# Function to adjust the curve based on historical data
def fit_rainbow_chart(df):
    df = df_final.copy()
    df["timestamp"] = pd.to_datetime(df["timestamp"])

    # Creating X and Y values ​​for curve adjustment
    xdata = np.arange(len(df)) + 1
    ydata = np.log(df["close"])  # We use log of values ​​to model the curve

    # Fitting the logarithmic curve
    popt, _ = curve_fit(log_func, xdata, ydata)

    return popt  # Returns the curve coefficients

# Function to classify each price on the Rainbow Chart
def classify_rainbow_chart(price, x, popt):
    base_value = np.exp(log_func(x, *popt))  # Base value expected by the model

    # Multipliers adjusted based on real model
    bands = {
        "🔥 Fire Sale": base_value * 0.4,
        "🟢 Buy!": base_value * 0.7,
        "🟡 Hold": base_value * 1.2,
        "🟠 Sell": base_value * 2,
        "🔴 FOMO": base_value * 3,
        "🚀 Super Bubble!": base_value * 5,
    }

    for label, threshold in bands.items():
        if price <= threshold:
            return label
    return "🚀 Super Bubble!"

# Main function to process the data and classify
def process_bitcoin_rainbow(file_path):
    df = pd.read_csv(file_path)
    df["timestamp"] = pd.to_datetime(df["timestamp"])

    # Adjust the curve based on historical data
    popt = fit_rainbow_chart(df)

    # Sort each record
    df["rainbow_class"] = df.apply(lambda row: classify_rainbow_chart(row["close"], row.name, popt), axis=1)

    return df

# Load and process data
df_rainbow = process_bitcoin_rainbow("/content/bitcoin_2013_until_2025.csv")
df_rainbow.to_csv("btc_rainbow_chart.csv", index=False)

# Display the latest records
print(df_rainbow.tail(10))
```

```python
import pandas as pd

def calculate_rsi(df, column="close", period=14):
    """
    Calculates the RSI (Relative Strength Index) for a specific column of a DataFrame.

    Parameters:
        df (pd.DataFrame): DataFrame containing closing prices.
        column (str): Name of the column that contains the prices. Default: "close".
        period (int): Period of the RSI. Default: 14.

    Returns:
        pd.Series: Calculated RSI.
    """
    df = df.copy()

    # Calculates price variation
    delta = df[column].diff()

    # Separate gains and losses
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    # Exponential moving average (EMA) of gains and losses
    avg_gain = gain.ewm(span=period, adjust=False).mean()
    avg_loss = loss.ewm(span=period, adjust=False).mean()

    # Calculation of RS and RSI
    rs = avg_gain / avg_loss
    df["RSI"] = 100 - (100 / (1 + rs))

    return df["RSI"]

# Usage example
df = df_rainbow.copy()
df["RSI"] = calculate_rsi(df, column="close", period=14)  # Calculate RSI
print(df.tail(10))  # Show last values
```
